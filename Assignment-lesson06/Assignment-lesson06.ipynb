# 使用iris（鸢尾花）数据集进行分类练习

```python
#导入iris数据集
from sklearn.datasets import load_iris 
data = load_iris()
data
```

```python
X,y,target = data['data'],data['target'],data['target_names']
X,y
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'X':X.tolist(),'y':y.tolist()})
df
```

```python
#将数据集随机分成训练数据和测试数据
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
#调用 knn 模型的 KNeighborsClassifier类
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=11) #n_neighbors 就是 k 个点，取距离最近的11个点，即 K = 25（可调整）
```

```python
knn.fit(X_train,y_train) #用  fit 方法对训练数据进行学习
```

```python
import numpy as np
import random
```

```python
#随机生成新数据进行测试
#X_new = np.random.rand(1,4)*10
X_new = np.random.randint(1,10,(1,4))
X_new
```

```python
# 用predict方法对新数据进行分类预测
prediction = knn.predict(X_new)

#预测结果
print("Predicted target name : {}".format(target[prediction]))
prediction
```

```python
#用测试数据对模型的准确性进行评估，准确率为100%。准确率与 k 的个数有关，k点的个数越多，准确率越高。
y_predict = knn.predict(X_test)

print(np.mean(y_predict == y_test))
```
